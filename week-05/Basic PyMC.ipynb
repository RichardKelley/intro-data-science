## Example - Flipping Coins

Suppose I'm tossing a potentially biased coin. The probability $p$ that the coin comes up heads is fixed, and we want to figure out a "good" guess for what $p$ is. What's a reasonable procedure for doing this?

Below, we look at the Bayesian approach.

```python
figsize(11, 9)

n_trials = [0, 1, 2, 3, 4, 5, 8, 15, 50, 500]
data = stats.bernoulli.rvs(0.5, size=n_trials[-1])
x = np.linspace(0, 1, 100)

for k, N in enumerate(n_trials):
    sx = plt.subplot(len(n_trials)/2, 2, k+1)
    xlabel("$p$, probability of heads") \
        if k in [0, len(n_trials)-1] else None
    setp(sx.get_yticklabels(), visible=False)
    heads = data[:N].sum()
    y = stats.beta.pdf(x, 1 + heads, 1 + N - heads)
    plot(x, y, label="observe %d tosses,\n %d heads" % (N, heads))
    fill_between(x, 0, y, color="#348ABD", alpha=0.4)
    vlines(0.5, 0, 4, color="k", linestyles="--", lw=1)

    leg = plt.legend()
    leg.get_frame().set_alpha(0.4)
    autoscale(tight=True)


suptitle("Bayesian updating of posterior probabilities",
             y=1.02,
             fontsize=14)

tight_layout()
```

Here, we start with a "prior belief" about the value of $p$, and as we observe more data, we update our belief to reflect that data. The updated belief is represented via a _posterior distribution_ over $p$. The really important thing to notice is that we're putting a probability distribution _on $p$_. We're treating the parameter of our random variable as if it were itself a random variable.

## Poisson Random Variables

We use these to count events in a fixed interval of time. The pmf is given by:

\\[P(X = k) = \frac{\lambda^k e^{-k}}{k!}, k \in \\{0,1,\ldots\\}\\]

And $E[X] = \lambda$.

```python
x = np.linspace(0,1,100)

figsize(12.5, 4)

import scipy.stats as stats
a = np.arange(16)
poi = stats.poisson
lambda_ = [1.5, 4.75]
colours = ["#348ABD", "#A60628"]

bar(a, poi.pmf(a, lambda_[0]), color=colours[0],
        label="$\lambda = %.1f$" % lambda_[0], alpha=0.60,
        edgecolor=colours[0], lw="3")

bar(a, poi.pmf(a, lambda_[1]), color=colours[1],
        label="$\lambda = %.1f$" % lambda_[1], alpha=0.60,
        edgecolor=colours[1], lw="3")

xticks(a + 0.4, a)
legend()
ylabel("probability of $k$")
xlabel("$k$")
title("Probability mass function of a Poisson random variable; differing \
$\lambda$ values")
```

## Exponential Random Variables

An _exponential_ random variable describes the time between events in a Poisson process. Its values can range from $0$ to $\infty$.

The density function is $f(x \mid \lambda) = \lambda \exp(-\lambda x)$ for $x \ge 0$, and $f(x \mid \lambda) = 0$ for $x < 0$.

If $X \sim \mbox{exponential}(\lambda)$, then $E[X] = \frac{1}{\lambda}$.

```python
a = np.linspace(0, 4, 100)
expo = stats.expon
lambda_ = [0.5, 1]

for l, c in zip(lambda_, colours):
    plot(a, expo.pdf(a, scale=1./l), lw=3,
             color=c, label="$\lambda = %.1f$" % l)
    fill_between(a, expo.pdf(a, scale=1./l), color=c, alpha=.33)

legend()
ylabel("PDF at $z$")
xlabel("$z$")
ylim(0,1.2)
title("Probability density function of an Exponential random variable;\
 differing $\lambda$");
```

## Example - Text Messages

This example uses actual texting frequency to ask an interesting question - does the user's texting behavior change over time? It's not obvious from the histogram, but the answer is yes. We're going to use PyMC to see that that is the case.

```python
figsize(12.5, 3.5)
count_data = np.loadtxt("txtdata.csv")
n_count_data = len(count_data)
plt.bar(np.arange(n_count_data), count_data, color="#348ABD")
plt.xlabel("Time (days)")
plt.ylabel("count of text-msgs received")
plt.title("Did the user's texting habits change over time?")
plt.xlim(0, n_count_data);
```

```python
import pymc as pm

alpha = 1.0/count_data.mean()
lambda_1 = pm.Exponential("lambda_1", alpha)
lambda_2 = pm.Exponential("lambda_2", alpha)

tau = pm.DiscreteUniform("tau", lower=0, upper=n_count_data)
```

```python
print "Random output:", tau.random(), tau.random(), tau.random()
```

```python
@pm.deterministic
def lambda_(tau=tau, lambda_1=lambda_1, lambda_2=lambda_2):
    out = np.zeros(n_count_data)
    out[:tau] = lambda_1  # lambda before tau is lambda1
    out[tau:] = lambda_2  # lambda after (and including) tau is lambda2
    return out
```

```python
observation = pm.Poisson("obs", lambda_, value=count_data, observed=True)

model = pm.Model([observation, lambda_1, lambda_2, tau])
```

```python
mcmc = pm.MCMC(model)
mcmc.sample(40000, 10000, 1)
```

```python
lambda_1_samples = mcmc.trace('lambda_1')[:]
lambda_2_samples = mcmc.trace('lambda_2')[:]
tau_samples = mcmc.trace('tau')[:]
```

```python
figsize(12.5, 10)
#histogram of the samples:

ax = subplot(311)
ax.set_autoscaley_on(False)

hist(lambda_1_samples, histtype='stepfilled', bins=30, alpha=0.85,
         label="posterior of $\lambda_1$", color="#A60628", normed=True)
legend(loc="upper left")
title(r"""Posterior distributions of the variables
    $\lambda_1,\;\lambda_2,\;\tau$""")
xlim([15, 30])
xlabel("$\lambda_1$ value")

ax = subplot(312)
ax.set_autoscaley_on(False)
hist(lambda_2_samples, histtype='stepfilled', bins=30, alpha=0.85,
         label="posterior of $\lambda_2$", color="#7A68A6", normed=True)
legend(loc="upper left")
xlim([15, 30])
xlabel("$\lambda_2$ value")

subplot(313)
w = 1.0 / tau_samples.shape[0] * np.ones_like(tau_samples)
hist(tau_samples, bins=n_count_data, alpha=1,
         label=r"posterior of $\tau$",
         color="#467821", weights=w, rwidth=2.)
xticks(np.arange(n_count_data))

legend(loc="upper left")
ylim([0, .75])
xlim([35, len(count_data)-20])
xlabel(r"$\tau$ (in days)")
ylabel("probability");
```
